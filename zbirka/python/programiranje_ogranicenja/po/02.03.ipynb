```python
import constraint
```

```python
problem = constraint.Problem()
```

```python
problem.addVariable('1_din', range(51))
problem.addVariable('2_din', range(26))
problem.addVariable('5_din', range(11))
problem.addVariable('10_din', range(6))
problem.addVariable('20_din', range(3))
```

```python
def f(d1, d2, d5, d10, d20):
    return 1*d1 + 2*d2 + 5*d5 + 10*d10 + 20*d20 == 50
```

```python
problem.addConstraint(f, ['1_din', '2_din', '5_din', '10_din', '20_din'])

# problem.addConstraint(constraint.ExactSumConstraint(50,[1, 2, 5, 10, 20]), ['1_din', '2_din', '5_din', '10_din', '20_din'])
```

```python
results = problem.getSolutions()

for result in results:
    # print(f"1 din: {result['1_din']}, 2 din: {result['2_din']}, 5 din: {result['5_din']}, 10 din: {result['10_din']}, 20 din: {result['20_din']}")
    print(result)
```
