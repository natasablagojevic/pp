```python
import constraint 
```

```python
problem = constraint.Problem()
```

```python
#  a      A 
#   b   B 
#     C 
#   d   D 
# e       E 
```

```python
problem.addVariables('abdeABCDE', range(1, 10))
```

```python
def f(a, b, d, e, C):
    return a + b + C + d + e == 25 

def f2(A, B, C, D, E):
    return A + B + C + D + E == 25
```

```python
problem.addConstraint(constraint.AllDifferentConstraint())
problem.addConstraint(f, ['a', 'b', 'd', 'e', 'C'])
problem.addConstraint(f2, ['A', 'B', 'C', 'D', 'E'])
```

```python
results = problem.getSolutions()

for result in results:
    print(result)
```
