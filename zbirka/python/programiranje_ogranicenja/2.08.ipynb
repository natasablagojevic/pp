```python
import constraint 
```

```python
#   A   B   C   D   E
#   F   G   H   I   J
#   K   L   M   N   O
#   P   K   R   S   T
#   U   V   W   X   Y
```

```python
problem = constraint.Problem()

problem.addVariables('AGMSYEIKU', range(1, 10))

def evaluate_sum(a, b, c, d, e):
    return a + b + c + d + e == 25

problem.addConstraint(constraint.AllDifferentConstraint())
problem.addConstraint(evaluate_sum, ['A', 'G', 'M', 'S', 'Y'])
problem.addConstraint(evaluate_sum, ['E', 'I', 'M', 'K', 'U'])

if problem.getSolutions() == []:
    print('Nema ressenja')
else:
    for result in problem.getSolutions():
        print('{} {} {} {} {}'.format(result['A'], '', '', '', result['E']))
        print('{} {} {} {} {}'.format('', result['G'], '', result['I'], ''))
        print('{} {} {} {} {}'.format('', '', result['M'], '', ''))
        print('{} {} {} {} {}'.format('', result['K'], '', result['S'], ''))
        print('{} {} {} {} {}'.format(result['U'], '', '', '', result['Y']))
        print()
```
